### I started with the cleaning the data ###


```python
Aviation_df.columns = Aviation_df.columns.str.replace('.', '_', regex=False)
```

```python
Aviation_df.columns
```

```python
# Convert `Event.Date` and `Publication.Date` to datetime
Aviation_df['Event_Date'] = pd.to_datetime(Aviation_df['Event_Date'], errors='coerce')
Aviation_df['Publication_Date'] = pd.to_datetime(Aviation_df['Publication_Date'], format='%d-%m-%Y', errors='coerce')
```

```python
# Convert `Latitude` and `Longitude` to numeric, coercing errors to NaN
Aviation_df['Latitude'] = pd.to_numeric(Aviation_df['Latitude'], errors='coerce')
Aviation_df['Longitude'] = pd.to_numeric(Aviation_df['Longitude'], errors='coerce')
```

```python
Aviation_df.head()
```

```python
# Fill missing numerical columns with the median
numerical_cols = ['Number_of_Engines', 'Total_Fatal_Injuries', 'Total_Serious_Injuries', 'Total_Minor_Injuries', 'Total_Uninjured']
for col in numerical_cols:
    if col in Aviation_df.columns:
        Aviation_df[col] = Aviation_df[col].fillna(Aviation_df[col].median())
```

```python
# Fill missing categorical columns with the mode
categorical_cols = [
    'Event_Id', 'Accident_Number', 'Location', 'Country', 'Airport_Code',
    'Airport_Name', 'Injury_Severity', 'Aircraft_damage', 'Aircraft_Category',
    'Registration_Number', 'Make', 'Model', 'Amateur_Built', 'Engine_Type',
    'FAR_Description', 'Purpose_of_flight', 'Weather_Condition',
    'Broad_phase_of_flight', 'Report_Status'
]
for col in categorical_cols:
    if col in Aviation_df.columns:
        Aviation_df[col] = Aviation_df[col].fillna(Aviation_df[col].mode()[0])
```

```python
# Fill missing Date columns with backfill
date_cols = ['Event_Date', 'Publication_Date']
for col in date_cols:
    if col in Aviation_df.columns:
        Aviation_df[col] = Aviation_df[col].bfill()
```

```python
#remove duplicates
Aviation_df.drop_duplicates(inplace=True)
```

```python
missing_percent = Aviation_df.isnull().mean().sort_values(ascending=False) * 100
missing_percent
```

```python
# Drop columns with more than 80% missing values

missing_percentage = Aviation_df.isnull().mean() * 100

columns_to_drop = missing_percentage[missing_percentage > 80].index.tolist()
Aviation_df_cleaned = Aviation_df.drop(columns=columns_to_drop)
```

##  Aviation Accident Data Analysis â€“ Problem Statements

This analysis explores aviation accident data to identify trends, patterns, and potential safety gaps in global air travel. The goal is to answer key investigative questions and derive actionable insights. Below are the major problem statements guiding this project:

---

### 1. What are the trends in aviation accidents over time?
- Are accidents increasing or decreasing over the years?
- Which years recorded the highest number of accidents?

---

### 2. Which locations are most prone to aviation accidents?
- Which countries or regions experience the most aviation incidents?
- Can we identify hotspots or high-risk areas on the map?

---

### 3. What are the most common causes or phases of flight where accidents occur?
- Do accidents mostly happen during takeoff, cruise, or landing?
- Which weather conditions are linked to more incidents?

---

### 4. Are certain aircraft models or manufacturers involved in more accidents?
- Which makes or models have the highest number of recorded accidents?
- Do amateur-built or older aircraft have higher accident rates?

---

### 5. What are the most severe or fatal aviation accidents?
- What is the death rate across different incidents?
- Which accidents had the highest number of fatalities?

---

### 6. Are certain flight purposes or engine types more at risk?
- Do commercial, private, or cargo flights experience more accidents?
- How do single-engine vs multi-engine aircraft compare in safety?


-These problem statements will guide our data cleaning, exploration, and visualization steps moving forward.



```python
# Re-categorize 'Injury_Severity' for better readability
def categorize_injury_severity(severity):
    if 'Fatal' in severity:
        return 'Fatal'
    elif 'Serious' in severity:
        return 'Serious'
    elif 'Minor' in severity:
        return 'Minor'
    elif 'Non-Fatal' in severity:
        return 'Non-Fatal'
    elif 'Incident' in severity:
        return 'Incident'
    else:
        return 'Other/Unknown'

Aviation_df_cleaned['Injury_Severity_Categorized'] = Aviation_df_cleaned['Injury_Severity'].apply(categorize_injury_severity)
```

```python
# Visualizing temporal trends using Event_Date and Publication_Date.
import matplotlib.pyplot as plt
import altair as alt
```

```python
# Extract the year like this:
Aviation_df_cleaned['Event_Date'] = pd.to_datetime(Aviation_df_cleaned['Event_Date'], errors='coerce')
Aviation_df_cleaned['Year'] = Aviation_df_cleaned['Event_Date'].dt.year

```

```python
# Count the number of events per year
events_per_year = Aviation_df_cleaned['Year'].value_counts().sort_index().reset_index()
events_per_year.columns = ['Year', 'Event_Count']
```

```python
# Plotting Events per Year (Event_Date)

#The chart below illustrates the number of aviation incidents and accidents based on their Event_Date.
chart_events = alt.Chart(events_per_year).mark_line(point=True).encode(
    x=alt.X('Year:O', axis=alt.Axis(title='Year')),
    y=alt.Y('Event_Count:Q', axis=alt.Axis(title='Number of Events')),
    tooltip=['Year', 'Event_Count']
).properties(
    title='Number of Aviation Events per Year (Event Date)'
).interactive()

chart_events
```

```python
# Extract the year from 'Event_Date' and 'Publication_Date' for plotting
Aviation_df_cleaned['Event_Year'] = Aviation_df_cleaned['Event_Date'].dt.year
Aviation_df_cleaned['Publication_Year'] = Aviation_df_cleaned['Publication_Date'].dt.year
```

```python
# Count the number of publications per year
publications_per_year = Aviation_df_cleaned['Publication_Year'].value_counts().sort_index().reset_index()
publications_per_year.columns = ['Year', 'Publication_Count']
```

```python
# Plotting Publications per Year (Publication_Date)

#The chart below displays the number of aviation incident and accident reports based on their Publication_Date.

chart_publications = alt.Chart(publications_per_year).mark_line(point=True).encode(
    x=alt.X('Year:O', axis=alt.Axis(title='Year')),
    y=alt.Y('Publication_Count:Q', axis=alt.Axis(title='Number of Publications')),
    tooltip=['Year', 'Publication_Count']
).properties(
    title='Number of Aviation Reports Published per Year (Publication Date)'
).interactive()

chart_publications
```

### Visualize 

```python

flight_phase_counts = (
    Aviation_df_cleaned['Broad_phase_of_flight']
    .value_counts()
    .sort_values(ascending=False)
)


plt.figure(figsize=(12, 6))
sns.barplot(
    x=flight_phase_counts.values,
    y=flight_phase_counts.index,
    hue=flight_phase_counts.index,  # Assign y variable to hue
    palette='coolwarm',
    dodge=False,
    legend=False
)
plt.title('Accidents by Flight Phase')
plt.xlabel('Number of Accidents')
plt.ylabel('Flight Phase')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

# Weather condition during accidents
print("\n Accidents by Weather Condition:\n")
print(Aviation_df_cleaned['Weather_Condition'].value_counts(dropna=False))
```

```python
#Group by Year and Weather_Condition
accidents_by_weather = (
    Aviation_df_cleaned
    .groupby(['Year', 'Weather_Condition'])
    .size()
    .unstack(fill_value=0)  # Turns Weather_Condition values into columns
    .sort_index()
)
```

### Visualize 

```python
# Plot All Accidents, IMC, and VMC Over Time
plt.figure(figsize=(14, 7))

# Plot each line
sns.lineplot(x=accidents_by_weather.index, y=accidents_by_weather.sum(axis=1), label='Total Accidents', marker='o', linewidth=2)
if 'IMC' in accidents_by_weather.columns:
    sns.lineplot(x=accidents_by_weather.index, y=accidents_by_weather['IMC'], label='IMC Accidents', marker='o')
if 'VMC' in accidents_by_weather.columns:
    sns.lineplot(x=accidents_by_weather.index, y=accidents_by_weather['VMC'], label='VMC Accidents', marker='o')

# Formatting
plt.title('Aviation Accident Trends by Weather Condition (IMC vs VMC)', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.legend()
plt.grid(True)

# Annotate global events
highlight_events = {
    2001: "9/11 Attacks",
    2008: "Global Financial Crisis",
    2020: "COVID-19"
}
for year, label in highlight_events.items():
    if year in accidents_by_weather.index:
        plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)
        plt.text(year + 0.2, plt.ylim()[1]*0.85, label, rotation=90, color='gray')

plt.tight_layout()
plt.show()
```
